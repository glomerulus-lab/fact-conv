```python
import numpy as np
import matplotlib.pyplot as plt
from estimator import classical_weights, V1_inspired_weights, haltere_inspired_weights
from palettable.colorbrewer.diverging import RdBu_11
from palettable.colorbrewer.sequential import Greys_9
```

# Image example

```python
M, N, t, l = 9, 100, 1, 2
dim = int(np.sqrt(N))
W = V1_inspired_weights(M, N, t, l, scale=1, random_state=7)

fig = plt.figure(figsize=(8, 6))
for i in range(3):
    ax = fig.add_subplot(1, 3, i + 1)
    plt.imshow(np.abs(W[i]).reshape(dim, dim), cmap=Greys_9.mpl_colormap)
    cmap_lim = np.max(np.abs(W[i]))
#     plt.clim(-cmap_lim, cmap_lim)
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
cbar = plt.colorbar(extend='neither', spacing='proportional',
                orientation='vertical', aspect=8)
cbar.ax.tick_params(axis='y', which='both', length=0, labelright=False)
plt.subplots_adjust(wspace=0.1, hspace=0)
```

```python
fig = plt.figure(figsize=(8, 6))
W = classical_weights(M, N, random_state=8)
for i in range(3):
    ax = fig.add_subplot(1, 3, i + 1)
    plt.imshow(W[i].reshape(dim, dim), cmap=Greys_9.mpl_colormap)
    cmap_lim = np.max(np.abs(W[i]))
    plt.clim(-cmap_lim, cmap_lim)
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
cbar = plt.colorbar(extend='neither', spacing='proportional',
                orientation='vertical', aspect=8)
cbar.ax.tick_params(axis='y', which='both', length=0, labelright=False)
plt.subplots_adjust(wspace=0.1, hspace=0)
```
